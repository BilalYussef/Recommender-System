```python
# Assigning dicstionary 
interests_dict = {'y': 'Devoloper',
'w': 'Data Science',
'j': 'basketball',
'c': 'Games',
't': 'Hicking',
'b': 'Music',
'a': 'Coffee',
'g': 'Entrepreneurs',
'p': 'Vollyball',
'h': 'Football',
'd': 'Anime',
'z': 'Moveies',
'l': 'Gym',
'i': 'Diet',
'k': 'Self development',
'v': 'History',
'u': 'Religon',
's': 'Travel',
'e': 'Diving',
'q': 'Swimming',
'm': 'Fishing',
'f': 'Hunting',
'o': 'Cars',
'n': 'Techenology',
'r': 'Food',
'x': 'Phtoghraphy'}
interests_list = list(interests_dict.values())
```

```python
# map interests_dict 
data.Interest = data.Interest.map(interests_dict)
```

```python
# applay random so we can generate random communities
import random

data['Interest2']= ''
data['Interest3']= ''
data['Interest4']= ''
data['Interest5']= ''

for i in range(200):
    random.seed(i)
    data.loc[i,['Interest','Interest2','Interest3','Interest4','Interest5']] = random.sample(interests_list, 5)
```

```python
# combian communities 

cols = ['Interest','Interest2','Interest3','Interest4','Interest5']
data['combined'] = data[cols].apply(lambda row: ', '.join(row.values.astype(str)), axis=1)
```

```python
data.sample(10)
```

```python
# implement recommanding system (C)
```

```python
from sklearn.metrics.pairwise import linear_kernel
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
community_tfidf = TfidfVectorizer()
```

```python
community_description_matrix = community_tfidf.fit_transform(data['combined'])
```

```python
community_description_matrix.shape
```

```python
community_description_matrix.toarray()
```

```python
cosine_similarity = linear_kernel(community_description_matrix, community_description_matrix)
```

```python
indices = pd.Series(data[['Interest','Interest2','Interest3','Interest4','Interest5']].index)
```

```python
def recommend(index, cosine_sim=cosine_similarity):
    id = indices[index]
    
    # Get the pairwsie similarity scores of all users 
    # sorting them and getting top 5
    
    similarity_scores = list(enumerate(cosine_sim[id]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similarity_scores = similarity_scores[1:6]

    # Get the users index
    community_index = [i[0] for i in similarity_scores]

    # Return the top 5 most similar user using integer-location based indexing (iloc)
    return data.iloc[community_index]
```

```python
recommend(10)
```

```python
n_users = merge_df.user_id.nunique()
n_community = merge_df.communityID.nunique()

print('Num. of Users: '+ str(n_users))
print('Num of Community: '+str(n_community))
```

```python
# communities have highest rate average 
ratings = pd.DataFrame(merge_df.groupby('combined')['community_ratings'].mean())
ratings.head()
```

```python
# communities have highest rate count
ratings['num of ratings'] = pd.Series(merge_df.groupby('combined')['community_ratings'].count())
ratings.head()
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
moviemat = merge_df.pivot_table(index='communityID', columns='combined', values='community_ratings')
moviemat.head()
```

```python
# Get rates for 'Star Wars (1977)' movie
community_user_ratings = moviemat['Entrepreneurs, Diving, Travel, Hicking, Moveies']
community_user_ratings.head()
```

```python
# get correlaion matrix 
similar_to_ = moviemat.corrwith(community_user_ratings)
corr_ = pd.DataFrame(similar_to_,columns=['Correlation'])
corr_.dropna(inplace=True)
corr_.head()
```

```python
corr_.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_ = corr_.join(ratings['num of ratings'])
corr_.sort_values('Correlation',ascending=False).head(10)
```
